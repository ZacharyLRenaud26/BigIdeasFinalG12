```python
import csv

def find_max_in_column(csv_file, column_name, neighborhood_column):
    max_value = None
    max_neighborhood = None

    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            value = float(row[column_name])  
            if max_value is None or value > max_value:
                max_value = value
                max_neighborhood = row[neighborhood_column]

    return max_value, max_neighborhood

csv_file = '61ab4ad4-fb84-4789-95c9-cbe116414287.csv'
column_name = 'bivalent_booster'
neighborhood_column = 'neighborhood_municipality'
max_value_in_column, max_neighborhood = find_max_in_column(csv_file, column_name, neighborhood_column)
print(f"The maximum value in column '{column_name}' is {max_value_in_column} in the neighborhood '{max_neighborhood}'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('61ab4ad4-fb84-4789-95c9-cbe116414287.csv')

sums = df.groupby(df["neighborhood_municipality"])["bivalent_booster"].sum()

sorted_sums = sums.sort_values(ascending=False)

threshold = sorted_sums.quantile(0.90)

top_10 = sorted_sums[sorted_sums >= threshold]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_10, labels=top_10.index, autopct='%1.1f%%', startangle=140, labeldistance=1.14)
plt.title('Top 10% Distribution of Maximum Values of bivalent_booster Across Neighborhoods', pad=30)
plt.axis('equal')
plt.show()
```
